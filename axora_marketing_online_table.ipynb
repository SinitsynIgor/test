```python
# выбираем нужные столбцы
df = data2.loc[:, ['Тариф', 'Маркетолог', 'Сайт', 'Axora фикс', 'Sale', 'Axora %', 'Axora %.1', 
                   'Axora %.2', '% Обслуж']][data2['Сумма']>0]
df
```

```python
# переименуем столбцы
df.rename(columns={'Тариф': 'tariff', 'Маркетолог':'marketolog',
                  'Сайт':'site', 'Axora фикс':'sale', 'Sale':'saleman', 'Axora %':'percent_ya',
                  'Axora %.1':'percent_g', 'Axora %.2':'percent_target', '% Обслуж':'percent_service'}, inplace=True)
df2 = df
df2
```

```python
# создадим фрейм для продаж
df3 = df2.loc[:,['tariff', 'site', 'sale', 'saleman']]
df3
```

```python
# переименуем sale в маркетологов для дальнейшего объединения
df3.rename(columns={'saleman':'marketolog'}, inplace=True)
df3
```

```python
# удалим лишние столбцы
df2.drop(columns=['sale','saleman'], inplace=True)
df2
```

```python
# объединим df2 И df3 в итоговый фрейм
df4 = df2.append(df3).fillna(0)
df4
```

```python
# посмотрим, где есть отсутсвующие значения
df4.isnull().any()
```

```python
df4.dtypes
```

```python
#
df4['total'] = df4.eval('percent_g + percent_service + percent_target + percent_ya + sale')
df4
```

```python
df5 = df4[df4['total']>0]
df5
```

```python
df6 = df5['total'].groupby(df5['marketolog']).resample('MS').sum()
df6.head()
```

```python
df7 = df6.unstack()
df7.head()
```

```python
int(df7.loc['Маша', '2020-07-01'])
```

```python
revenue = {
    'Маша': df6['Маша'].sum(),
    'Лена': df6['Лена'].sum(),
    'Вячеслав': df6['Вячеслав'].sum(),
    'Полина': df6['Полина'].sum(),
    'Игорь': df6['Игорь'].sum(),
    'Ульяна': df6['Ульяна'].sum()
}
a = [
    ['Наготко', 'Мария', 'Маша'],
    ['Куделко', 'Елена', 'Лена'],
    ['Стаселько', 'Ульяна', 'Ульяна'],
    ['Коробко', 'Полина', 'Полина'],
    ['Синицын', 'Игорь', 'Игорь'],
    ['Мацкевич', 'Вячеслав', 'Вячеслав']
]

for surname, name, s in a:
    text = f"""Выручка специалиста {name} {surname} составила {int(revenue[s])} р."""
    print(text)
```
